```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = range(100)
y = range(100)

plt.plot(x, y)
plt.show()
```

##  matplotlib 

- 최대 단점 argument를 kwargs를 받음
    - plt.plot(*kwargs)
- 고정된 argument가 없어서 alt + tab으로 확인이 어려움
- Graph는 원래 figure객체에 생성
- pyplot 객체 사용시, 기본 figure에 그래프 그려짐

```python
x_1 = range(100)
y_1 = [np.cos(value) for value in x_1]

x_2 = range(100)
y_2 = [np.sin(value) for value in x_2]

plt.plot(x_1, y_1)
plt.plot(x_2, y_2)
plt.show()
```

## Figure & Axes
- matplotlib는 Figure 안에 여러개의 Axes로 구성
- Figure 위에 여러개의 Axes를 생성

```python
fig = plt.figure() # figure를 반환
fig.set_size_inches(10, 5) # 크기지정
ax_1 = fig.add_subplot(1, 2, 1) # 1*2 크기의 figure에 1번째 구역에 plot하겠다.
ax_2 = fig.add_subplot(1, 2, 2) # 1*2 크기의 figure에 2번째 구역에 plot하겠다.

ax_1.plot(x_1, y_1, c='b')
ax_2.plot(x_2, y_2, c='g')
plt.show()
```

##  subplots
- subplot 순서 결정
- subplot(세로개수, 가로개수, 순서)

```python
fig = plt.figure()
fig.set_size_inches(10, 10)

ax_1 = fig.add_subplot(2, 2, 1)
ax_2 = fig.add_subplot(2, 2, 2)
ax_3 = fig.add_subplot(2, 1, 2)

ax_1.plot(x_1, y_1, c='b')
ax_2.plot(x_2, y_2, c='g')
ax_3.plot(x_1, y_1)

plt.show()
```

```python
# 한번에 여러가지 plot을 handling
fig = plt.figure()
fig.set_size_inches(10, 10)
plt.style.use('ggplot')

ax = []
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
for i in range(1, 7):
    ax.append(fig.add_subplot(2, 3, i))
    x_1 = np.arange(50)
    y_1 = np.random.randn(50)
    c = colors[np.random.randint(1, len(colors))]
    
    ax[i-1].plot(x_1, y_1, c = c)
```

```python
# subplot properties
fig = plt.figure()
fig.set_size_inches(10, 10)

ax_1 = fig.add_subplot(2, 2, 1)
ax_2 = fig.add_subplot(2, 2, 2)
ax_3 = fig.add_subplot(2, 1, 2)

ax_1.plot(x_1, y_1, c='b')
ax_2.plot(x_2, y_2, c='g')
ax_3.plot(x_1, y_1)

ax_2.margins(2, 2) #Values >0.0 zoon out

left = 0.125 # the left side of the subplots of the figure
right = 0.9 # the right side of the subplots of the figure
bottom = 0.1 # the bottom of the subplots of the figure
top = 0.9 # the top of the subplots of the figure
wspace = 0.5 # the amount of width reserved for blank space between subplots
hspace = 0.1 # the amount of width reserved for white space between subplots

fig.subplots_adjust(left, bottom, right, top, wspace, hspace=None)

plt.show()
```

## Examples

###  월별 통화량 데이터 총합 구해보기

```python
df = pd.read_csv('./data/phone_data.csv')
df.head()
```

```python
duration = df.groupby(['month'])['duration'].sum()
duration
```

```python
plt.plot(duration.index, duration, color = 'b', linestyle = 'dashed')
plt.title('Durations per month')
plt.show()
```

## Set color
- color 속성 사용
- float -> 흑백, rgb color, predefined color

```python
plt.plot(x_1, y_1, color ="#eeefff")
plt.plot(x_2, y_2, color = 'r')

plt.show()
```

## Set title

```python
plt.plot(x_1, y_1, c = 'b')
plt.plot(x_2, y_2, c = 'g')

plt.title("Twe lines")
plt.show()
```

## Set legend

```python
plt.plot(x_1, y_1, c = 'b', linestyle = 'dashed', label= 'line_1')
plt.plot(x_2, y_2, c = 'g', linestyle = 'dotted', label = 'line_2')
plt.legend(shadow = True, fancybox = True, loc = 'lower right')

plt.title("Twe lines")
plt.show()
```

```python
# label
plt.plot(x_1, y_1, c = 'b', linestyle = 'dashed', label= 'line_1')
plt.plot(x_2, y_2, c = 'g', linestyle = 'dotted', label = 'line_2')
plt.legend(shadow = True, fancybox = True, loc = 'lower right')

plt.title("Twe lines")
plt.xlabel('X_label')
plt.ylabel('Y_label')
plt.show()
```
