##  matplotlib 

- 최대 단점 argument를 kwargs를 받음

```python
x_1 = range(100)
y_1 = [np.cos(value) for value in x_1]

x_2 = range(100)
y_2 = [np.sin(value) for value in x_2]

plt.plot(x_1, y_1)
plt.plot(x_2, y_2)
plt.show()
```

## Figure & Axes
- matplotlib는 Figure 안에 여러개의 Axes로 구성
- Figure 위에 여러개의 Axes를 생성

```python
fig = plt.figure() # figure를 반환
fig.set_size_inches(10, 5) # 크기지정
ax_1 = fig.add_subplot(1, 2, 1) # 1*2 크기의 figure에 1번째 구역에 plot하겠다.
ax_2 = fig.add_subplot(1, 2, 2) # 1*2 크기의 figure에 2번째 구역에 plot하겠다.

ax_1.plot(x_1, y_1, c='b')
ax_2.plot(x_2, y_2, c='g')
plt.show()
```

##  subplots
- subplot 순서 결정
- subplot(세로개수, 가로개수, 순서)

```python
fig = plt.figure()
fig.set_size_inches(10, 10)

ax_1 = fig.add_subplot(2, 2, 1)
ax_2 = fig.add_subplot(2, 2, 2)
ax_3 = fig.add_subplot(2, 1, 2)

ax_1.plot(x_1, y_1, c='b')
ax_2.plot(x_2, y_2, c='g')
ax_3.plot(x_1, y_1)

plt.show()
```

```python
# 한번에 여러가지 plot을 handling
fig = plt.figure()
fig.set_size_inches(10, 10)
plt.style.use('ggplot')

ax = []
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
for i in range(1, 7):
    ax.append(fig.add_subplot(2, 3, i))
    x_1 = np.arange(50)
    y_1 = np.random.randn(50)
    c = colors[np.random.randint(1, len(colors))]
    
    ax[i-1].plot(x_1, y_1, c = c)
```

```python
# subplot properties
fig = plt.figure()
fig.set_size_inches(10, 10)

ax_1 = fig.add_subplot(2, 2, 1)
ax_2 = fig.add_subplot(2, 2, 2)
ax_3 = fig.add_subplot(2, 1, 2)

ax_1.plot(x_1, y_1, c='b')
ax_2.plot(x_2, y_2, c='g')
ax_3.plot(x_1, y_1)

ax_2.margins(2, 2) #Values >0.0 zoon out

left = 0.125 # the left side of the subplots of the figure
right = 0.9 # the right side of the subplots of the figure
bottom = 0.1 # the bottom of the subplots of the figure
top = 0.9 # the top of the subplots of the figure
wspace = 0.5 # the amount of width reserved for blank space between subplots
hspace = 0.1 # the amount of width reserved for white space between subplots

fig.subplots_adjust(left, bottom, right, top, wspace, hspace=None)

plt.show()
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)

colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N)) ** 2 #값에 따라 점의 크기를 다르게 지정
plt.scatter(x, y, s=area, c = colors, alpha = 0.5)
plt.show()
```

## Bar Chart
- bar함수 사용

```python
data = [[5., 25., 50., 20.], [4., 23., 51., 17.], [6., 22., 52., 19.]]

X = np.arange(4)
plt.bar(X + 0.00, data[0], color='b', width = 0.25)
plt.bar(X + 0.25, data[1], color='g', width = 0.25)
plt.bar(X + 0.50, data[2], color='r', width = 0.25)
plt.xticks(X + 0.25, ("A", 'B', 'C', 'D'))
# 눈금이 각 데이터의 가운데에 적히도록 하기위해 X + 0.25를 해준다
plt.show()
```

```python
# bar chart stack style

data = np.array([[5., 5., 5., 5.,], [4., 23., 51., 17.],
                 [6., 22., 52., 19.]])
color_list = ['b', 'r', 'g']
data_label = ['A', 'B', 'C']
X = np.arange(data.shape[1])

for i in range(data.shape[0]):
    plt.bar(X, data[i], bottom = np.sum(data[:i], axis = 0),
            color = color_list[i], label = data_label[i])
    # bottom을 기준으로 다음 데이터가 쌓을 위치를 제공
plt.legend()
plt.show()
```

```python
# bar chart horisontal style
women_pop = np.array([5, 30, 45, 22])
men_pop = np.array([5, 25, 50, 20])
X = np.arange(4)

plt.barh(X, women_pop, color = 'r')
plt.barh(X, -men_pop, color = 'g')
plt.show()
```

# Matplotlib with pandas
- matplotlib를 사용한 그래프 지원

```python
df_data = pd.read_csv('./data/housing.data', sep="\s+", header = None)
# csv타임 데이터 로드, seperate는 빈공간으로 지정하고, Column은 없음
df_data.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE',
             'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']

df_data.head()
```

```python
fig = plt.figure()

ax = []
for i in range(1, 5):
    ax.append(fig.add_subplot(2, 2, i))
    
ax[0].scatter(df_data['CRIM'], df_data['MEDV'], s = 3)
ax[1].scatter(df_data['PTRATIO'], df_data['MEDV'])
ax[2].scatter(df_data['AGE'], df_data['MEDV'])
ax[3].scatter(df_data['NOX'], df_data['MEDV'])
plt.show()
```

```python
# coding with loop

fig = plt.figure()
ax = []

for i in range(1, 5):
    ax.append(fig.add_subplot(2, 2, i))
    
columns = ["CRIM", "PTRATIO", "AGE", 'NOX']
colors = ['b', 'g', 'c', 'r']
for i in range(4):
    ax[i].scatter(df_data[columns[i]], df_data['MEDV'], c =colors[i], label = columns[i])
    ax[i].legend()
    ax[i].set_title(columns[i])

plt.subplots_adjust(wspace = 0, hspace = 0)
plt.show()
```

## Seaborn - multiple plots

###  predefined multiple plots
- replot : Numeric 데이터 중심의 분포 / 선형 표시
- catplot : category 데이터 중심의 표시
- FacetGrid : 특정 조건에 따른 다양한 plot을 grid로 표시
- pairplot : 데이터 간의 상관관계 표시
- lmplot : regression 모델과 category 데이터를 함께 표시

```python
sns.set(style ='white', palette='muted', color_codes=True)
rs = np.random.RandomState(10)

f, axes = plt.subplots(2, 2, figsize=(7, 7), sharex = True)
sns.despine(left=True)

d = rs.normal(size = 100)

sns.distplot(d, kde = False, color = 'b', ax = axes[0, 0])
sns.distplot(d, hist = False, rug = True, color = 'r', ax = axes[0, 1])
sns.distplot(d, hist = False, color = 'g', kde_kws = {'shade': True},
             ax = axes[1,0])
sns.distplot(d, color = 'm', ax = axes[1, 1])

plt.setp(axes, yticks=[])
plt.tight_layout()
```

### pairplot - 데이터 간의 상관관계 표시

```python
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
# diag 에만 다른 plot을 그릴 수있음
g.map_offdiag(plt.scatter)
```

```python
g = sns.PairGrid(iris, hue = 'species')
# 상관 관계 plot에서 카테고리 별로 plot을 구분하여 보여준다.
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
```
